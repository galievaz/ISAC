# Information retrieval


The preprocess function converts words in files.

removing '\n'
removing punctuation
filter the non stopwords
removing the stem

```python
def processing(file):
    parts=file.split('\n')
    sentences=copy.copy(parts)
    #print(parts)
    #print(len(parts))
    for i in range(len(parts)):
        #Tokenizing the string object from the text file
        tokens = word_tokenize(parts[i])
    
        #Removing punctuations and making all letters to lower case
        words = [w.lower() for w in tokens if w.isalpha()]
    
        #Removing the stopwords from the list of words
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [w for w in words if w not in stop_words]
    
        #Stemming the list of words
        porter = nltk.PorterStemmer()
        stemmed = [porter.stem(t) for t in filtered_tokens]
        parts[i]=filtered_tokens
        
    return parts,sentences
    
    
docs,sentences = processing(abstracts)
#print(docs)
```

After getting list of lists with tokens, we willl get inverted index dictionary, in which keys are unique terms and their values are indexes of documents containing these terms

```python
from collections import defaultdict

inv_indx = defaultdict(list)
for idx, text in enumerate(docs):
    for word in text:
        inv_indx[word].append(idx)
```

# Task 1

For getting shared word count we recall function CountFrequency for counting each value of keys, then dictionary with got values will sort and choose top 10 ranked documents and print them

```python
import heapq
def CountFrequency(my_list):
    freq = {} 
    #create dict with values as counters 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    #for key, value in freq.items(): 
        #print ("% d : % d"%(key, value))
        
    #sort 10 top biggest values
    ranked=heapq.nlargest(10, freq, key=freq.get)
    print(ranked)
        
def shared_word_count(query,inv_indx):
    query=query.lower()
    query=query.split(' ')
    similarity=0
    for i in range(len(query)):
        #print( query[i])
        if query[i] in inv_indx:
            w=1
            es = dict(filter(lambda item: query[i] in item[0] and len(item[0])== len(query[i]), inv_indx.items()))
            CountFrequency(es[query[i]])
        else:
            print("Word of query '",query[i],"' not found")
            w=0
        similarity=similarity+w  
    return similarity
query=input()
summa=shared_word_count(query, inv_indx)
print("Similarity = ", summa)
```

It's similar to task 1, only with adding bonus 1/df(j) to similarity, where df(j) is number of documents where word j occurs

```python
import heapq
def CountFrequency(my_list):
    freq = {} 
    #create dict with values as counters 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    #for key, value in freq.items(): 
        #print ("% d : % d"%(key, value))
        
    #sort 10 top biggest values
    
    ranked=heapq.nlargest(10, freq, key=freq.get)
    #print(ranked)
    return ranked, len(freq)
        
def shared_word_count(sentences,query,inv_indx):
    query=query.lower()
    query=query.split(' ')
    similarity=0
    for i in range(len(query)):
        #print( query[i])
        if query[i] in inv_indx:
            
            es = dict(filter(lambda item: query[i] in item[0] and len(item[0])== len(query[i]), inv_indx.items()))
            results, df=CountFrequency(es[query[i]])
            w=1+1/df
        else:
            print("Word of query '",query[i],"' not found")
            w=0
        similarity=similarity+w
        for i in range(len(results)):
            j=results[i]
            print("Result ", i, ", abstract in line #", j)
            print(sentences[j])   
    return  similarity
query=input()
summa=shared_word_count(sentences,query, inv_indx)
print("Similarity = ", summa)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def cosine_similarity(query, inv_indx,sentences):
    query= query.lower().split()
    l1=[]
    l2=[]
    array=[]
    sw=stopwords.words('english')
    for i in range(len(query)):
        #print( query[i])
        if query[i] in inv_indx:
            Y_set = {query[i]}   
            es = dict(filter(lambda item: query[i] in item[0] and len(item[0])== len(query[i]), inv_indx.items()))
            results, df=CountFrequency(es[query[i]])
            for j in range(len(results)):
                number=results[j]
                X=sentences[number]
                X_list = word_tokenize(X) 
                X_set = {w for w in X_list if not w in sw}  
                #print(X_set)
                #print(Y_set)
                rvector = X_set.union(Y_set)  
                for w in rvector: 
                    if w in X_set: l1.append(1) # create a vector 
                    else: l1.append(0) 
                    if w in Y_set: l2.append(1) 
                    else: l2.append(0) 
                c = 0
                # cosine formula  
                for k in range(len(rvector)): 
                    c+= l1[k]*l2[k] 
                    cosine = c / float((sum(l1)*sum(l2))**0.5) 
                print("similarity: ", cosine) 
                array.append(cosine)
query=input()
summa=cosine_similarity(query, inv_indx, sentences)
print("Similarity = ", summa)
```
